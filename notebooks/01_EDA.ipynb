```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Target distribution
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Late_delivery_risk', data=df, palette='Set2')
plt.title('Distribution of Late Delivery Risk')
plt.xlabel('Late Delivery Risk (1 = Late, 0 = On-time)')
plt.ylabel('Count')

# annotate counts + percentages on bars
total = len(df)
for p in ax.patches:
    count = int(p.get_height())
    pct = 100 * count / total
    ax.annotate(f'{count}\n({pct:.1f}%)', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.tight_layout()
plt.show()

print(df['Late_delivery_risk'].value_counts(normalize=True).round(3) * 100)
```

```python
categorical_cols = ['Shipping Mode', 'Order Region', 'Market', 'Customer Segment', 'Department Name']

fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(10, 5 * len(categorical_cols)))

for ax, col in zip(axes, categorical_cols):
    rate = df.groupby(col)['Late_delivery_risk'].mean().sort_values(ascending=False)
    rate.plot(kind='barh', ax=ax, color='darkorange')
    ax.set_title(f'Late Delivery Rate by {col}')
    ax.set_xlabel('Late Delivery Rate')
    ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
numeric_cols = [
    'Days for shipment (scheduled)',
    'Order Item Discount Rate',
    'Benefit per order',
    'Sales per customer'
]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for ax, col in zip(axes, numeric_cols):
    sns.boxplot(x='Late_delivery_risk', y=col, data=df, palette='Set2', ax=ax)
    ax.set_title(f'{col} by Late Delivery Risk')
    ax.set_xlabel('Late Delivery Risk (1 = Late, 0 = On-time)')
    ax.set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python
numeric_cols = [
    'Days for shipment (scheduled)',
    'Benefit per order',
    'Sales per customer',
    'Order Item Discount',
    'Order Item Discount Rate',
    'Order Item Product Price',
    'Order Item Profit Ratio',
    'Order Item Quantity',
    'Sales',
    'Order Item Total',
    'Order Profit Per Order',
    'Product Price',
    'Late_delivery_risk'
]

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 9))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Matrix — Numerical Features + Target')
plt.tight_layout()
plt.show()
```

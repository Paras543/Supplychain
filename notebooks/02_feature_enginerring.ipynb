```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
```

```python
df['order_month'] = df['order date (DateOrders)'].dt.month
df['order_dow'] = df['order date (DateOrders)'].dt.dayofweek  # 0 = Monday, 6 = Sunday
df['is_weekend'] = df['order_dow'].isin([5, 6]).astype(int)
```

```python
X = df.drop('Late_delivery_risk',axis=1)
```

```python
y = df['Late_delivery_risk']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape,X_test.shape
```

```python
id_cols = ['Order Id', 'Order Item Id', 'Order Customer Id', 'Order Item Cardprod Id']

X_train = X_train.drop(columns=id_cols)
X_test = X_test.drop(columns=id_cols)

print(X_train.shape, X_test.shape)
```

```python
df.nunique()
```

```python
object_cols = X_train.select_dtypes(include='object').columns.tolist()
print(object_cols)

cardinality = X_train[object_cols].nunique().sort_values(ascending=False)
print(cardinality)
```

```python
from sklearn.preprocessing import OneHotEncoder
low_card_cols = ['Type', 'Customer Segment', 'Customer Country', 'Market',
                  'Shipping Mode', 'Order Region', 'Department Name', 'Order Status']

ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
ohe.fit(X_train[low_card_cols])

## Transfrom the training data 
X_train_transform = pd.DataFrame(
    ohe.transform(X_train[low_card_cols]),
    columns=ohe.get_feature_names_out(low_card_cols),
    index = X_train.index

)

X_test_transform = pd.DataFrame(
    ohe.transform(X_test[low_card_cols]),
    columns=ohe.get_feature_names_out(low_card_cols),
    index = X_test.index

)
```

```python
print(X_train_transform.shape, X_test_transform.shape)
```

```python
high_card_cols = ['Order City', 'Order State', 'Customer City',
                   'Order Country', 'Product Name', 'Category Name', 'Customer State']

freq_maps = {}

for col in high_card_cols:
    freq_maps[col] = X_train[col].value_counts(normalize=True)  # proportion, not raw count

train_freq = pd.DataFrame(index=X_train.index)
test_freq = pd.DataFrame(index=X_test.index)

for col in high_card_cols:
    train_freq[col + '_freq'] = X_train[col].map(freq_maps[col])
    test_freq[col + '_freq'] = X_test[col].map(freq_maps[col]).fillna(0)

print(train_freq.shape, test_freq.shape)
print(train_freq.isnull().sum().sum(), test_freq.isnull().sum().sum())

```

```python
all_object_cols = low_card_cols + high_card_cols

X_train_final = pd.concat(
    [X_train.drop(columns=all_object_cols), X_train_transform, train_freq],
    axis=1
)

X_test_final = pd.concat(
    [X_test.drop(columns=all_object_cols), X_test_transform, test_freq],
    axis=1
)

print(X_train_final.shape, X_test_final.shape)
print(X_train_final.dtypes.value_counts())
```

```python
X_train_final = X_train_final.drop(columns=['order date (DateOrders)'])
X_test_final = X_test_final.drop(columns=['order date (DateOrders)'])

print(X_train_final.shape, X_test_final.shape)
print(X_train_final.dtypes.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(
    scaler.fit_transform(X_train_final),
    columns=X_train_final.columns,
    index = X_train_final.index
)
X_test_scaled = pd.DataFrame(
    scaler.transform(X_test_final),
    columns=X_test_final.columns,
    index = X_test_final.index
)
```

```python
print(X_train_scaled.shape, X_test_scaled.shape)
X_train_scaled.describe().loc[['mean', 'std']]
```

```python
import os

os.makedirs('../data/processed', exist_ok=True)

X_train_scaled.to_csv('../data/processed/X_train.csv', index=False)
X_test_scaled.to_csv('../data/processed/X_test.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)
y_test.to_csv('../data/processed/y_test.csv', index=False)

print("Saved all processed files")
```
